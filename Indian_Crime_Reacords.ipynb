```python
import pandas as pd
import matplotlib as plt

def murder_female(x):
    read = pd.read_csv('32_Murder_victim_age_sex.csv')
    read.head()
    read_total = read [(read['Year']==x) & (read['Sub_Group_Name']=='2. Female Victims')]

    #Plotting age breakup of victims
    ax = read_total.plot(kind='bar',legend=True, title = 'female victims of murder')
    ax.set_ylabel("no. of victims", fontsize=12)
    ax.set_xticklabels(read['Area_Name'])
```
